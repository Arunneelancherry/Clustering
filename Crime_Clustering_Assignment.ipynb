# **K-Means**

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
Crime = pd.read_csv('crime_data.csv')
Crime
```

```python
def minmax_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)
```

```python
x=minmax_func(Crime.iloc[:,1:])
x
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_Crime=scaler.fit_transform(Crime.iloc[:,1:])
scaled_Crime
```

```python
wcss=[]
for i in range (1,11):
  kmeans=KMeans(n_clusters=i,random_state=0)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)

  print(wcss)

plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS (Within Cluster Sum of Squares)')
plt.show()
```

```python
from sklearn.cluster import KMeans
cluster_new =KMeans(4,random_state=0)
cluster_new.fit(x)
```

```python
cluster_new.labels_
```

```python
Crime['Clusterid_New']=cluster_new.labels_
Crime
```

```python
cluster_new.cluster_centers_
```

```python
Crime.groupby('Clusterid_New').agg(['mean']).reset_index()
```

```python
Crime['Clusterid_New'].value_counts()
```

```python
import seaborn as sns
sns.scatterplot(data=Crime)
```

# DBSCAN

```python
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
```

```python
from google.colab import files
uploaded= files.upload()
```

```python
crime = pd.read_csv('crime_data.csv')
crime
```

```python
crime.info()
```

```python
crime.drop('Unnamed: 0',axis=1,inplace =True)
```

```python
data=crime.values
data
```

```python
stscaler=StandardScaler()
X=stscaler.fit_transform(data)
X
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
crime['Clusterid']=dbscan.labels_
crime
```

```python
crime['Clusterid'].value_counts()
```

```python
(len(crime[crime['Clusterid']==-1])/len(crime))*100
```

```python
X= metrics.silhouette_score(X,dbscan.labels_)
X
```

```python
crime.groupby(crime['Clusterid']).mean()
```
